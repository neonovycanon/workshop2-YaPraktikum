#### Выбор репрезентативного подмножества

Представленные заказчиком датасеты связаны друг с другом, так как являются подмножествами основного датасета `base`.

Анализ данных можно проводить на репрезентативной выборке, долю выборки определим путём расчёта критерия `Колмогорова-Смирнова` (уровень доверия - 0.95).

```python
def ks_tests(data:pd.DataFrame,
            frac: np.array,
            alpha = 0.05,
            random_state = 42):
    cols = data.select_dtypes(include=[np.number]).columns.tolist()
    test_res = {}
    for i in cols:
        temp = []
        for j in frac:
            stat, p_value = ks_2samp(data[i], data[i].sample(frac = j, random_state= random_state))
            if p_value >= alpha:
                temp.append(j)
            else:
                temp.append(np.nan)
        test_res.update({i:temp})
    return test_res
```

При расчёте тестов `Колмогорова-Смирнова` уровень доверия 0.95 ($\alpha = 0.05$):

- Нулевая гипотеза $H_0$: Выборка и основные данные имеют одинаковое распределение - $p_{value}\geq \alpha$;
- Гипотеза $H_1$: Выборка и основные данные не представляют одинаковое распределение - $p_{value}\leq \alpha$.

#### Визуализация распределений данных

Известно, что все данные представлены *непререрывными численными признаками*. Далее визуализируем распределения признаков с целью анализа и поиска информации в данных.

Для анализа дополнительно используем функцию оценки количества выбросов `find_outliers`, позволяющую определить количество выбросов путём расчёта границ $X_{mean} \pm 1.5\cdot{IQR}$, используя межквартильное расстояние.

```python
def find_outliers_IQR(x: pd.DataFrame,
                  n = 1.5):
    x = x.select_dtypes(include=np.number)
    outliers_count = {}
    for i in x.columns:
        j = 0
        x_25, x_75 = np.quantile(x[i], 0.25), np.quantile(x[i], 0.75)
        x_IQR = x_75 - x_25
        x_50 = np.mean(x[i])
        lb, ub = (x_50 - n*x_IQR), (x_50 + n*x_IQR)
        j = x[(x[i] < lb) | (x[i] > ub)].shape[0]
        outliers_count.update({i:j})
    return outliers_count
```

```python
def hist_plotting(data: pd.DataFrame,
                  columns: list,
                  out_liers: dict,
                  grid_size = (18, 4),
                  fig_s = (8, 6),
                  dpi = 120):
    
    fig, axs = plt.subplots(grid_size[0], grid_size[1],
                            constrained_layout = True,
                            figsize = fig_s,
                            dpi = dpi)
    plt.suptitle('Гистограммы распределений признаков')
    for i, j in enumerate(columns):
        x_c = i // grid_size[1]
        y_c = i % grid_size[1]
        pos = axs[x_c, y_c]
        sns.histplot(data[j], ax=pos)
        pos.set_title(f'Признак \n {j}')
        pos.set_xlabel('Значение')
        pos.set_ylabel('Наблюдения')
        pos.text(0.5, -0.5, 'Выбросы, ед.: '+str(out_liers.get(j)), transform=pos.transAxes, 
                 fontsize=10, verticalalignment='top', horizontalalignment='center',
                 bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))
```

Определив количество выбросов с помощью функции `find_outliers` далее воспользуемся функцией `hist_plotting` с целью построить распределения каждого из признаков.

### Решение задачи мэтчинга

Задачи по нахождению ближайших соседей являются классом задач `Обучения без учителя (Unsupervised learning)`. В таких типах задач алгоритмы используются для нахождения паттернов на неразмеченных данных. Существуют различные подходы к решению такого класса задач. 

Задача может быть решена методами кластеризации следующим образом:

- Поиск ближайших объектов методом `brute-force` - расчёт расстояний (зачастую норм: $L_2$ или $cos_{sim}$) по всем объектам выборки (сопровождается проклятием размерности);
- Поиск ближайших соседей на размеченных данных:
    - Данные разделяются на n - кластеров, каждому из которых присваивается в соответствие своя центроида. Способ разделения зависит от типа использованного алгоритма кластеризации;
    - Поиск ближайших кластеров (расчёт расстояний до центроид) и поиск внутри этих кластеров.

Для решения задачи мэтчинга используем бибилотеки: `FAISS` и `Scikit-learn`.

#### Алгоритм FAISS
Библиотека `FAISS` используется для эффективного поиска наиболее близких векторов и кластеризации `dense vectors` (плотных векторов). Основа библиотеки `FAISS` класс `Index`, для которого реализованы функции расчёта расстояний. Бибилотека работает на векторах фиксированной размерности, при этом также используются и методы **квантизации**, позволяющие ускорять расчёт расстояний.

Тестирование гипотез для бибилиотеки `FAISS` будем проводить на сжатом датасете `base_data_c`.

##### Baseline модель

В алгоритме реализуется поиск по данным, прошедшим *сегментацию*, т.е. сначала определяются наиболее близкие к запросу ячейки, далее происходит поиск внутри ячейки.

Поиск осуществялется с помощью индекса `IndexIVFFlat`, который требует *обучения* (обучение может происходить как на тренировочной выборке, так и на всём массиве данных).

Для работы индекса `IndexIVFFlat` требуется дополнительный индекс - *квантизатор* `IndexFlatL2`, присваивающий векторы ячейкам Вороного.

```python
#Настройка размерности векторов
#dims = base_data_c.shape[1] 
dims = base_data_f.shape[1]

#Настройка числа ячеек Вороного
n_cells = 20

#Квантизатор - присваивает объекты к ячейкам Вороного
quantizer = faiss.IndexFlatL2(dims)

#IndexIVFFlat доп. индекс, используемый для выделения сегментов датасета. ТРЕБУЕТ ОБУЧЕНИЯ.
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```
